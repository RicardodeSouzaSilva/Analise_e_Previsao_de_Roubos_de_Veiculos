```python
# Importa o módulo 'os' para interagir com o sistema operacional, como navegar por diretórios e listar arquivos
import os

# Importa a biblioteca pandas, que é amplamente utilizada para manipulação e análise de dados em Python
import pandas as pd

# Obtém o diretório de trabalho atual onde o script está sendo executado
diretorio = os.getcwd()

# Cria uma lista contendo os nomes dos arquivos no diretório atual que terminam com a extensão '.xlsx'
# Isso filtra apenas arquivos Excel
arquivos_xlsx = [arquivo for arquivo in os.listdir(diretorio) if arquivo.endswith('.xlsx')]

# Verifica se existem pelo menos dois arquivos XLSX para que a combinação de dados faça sentido
if len(arquivos_xlsx) >= 2:
    # Inicializa uma lista vazia para armazenar os DataFrames que serão criados a partir de cada arquivo Excel
    dataframes = []

    # Loop por cada arquivo XLSX identificado no diretório
    for caminho_arquivo in arquivos_xlsx:
        # Lê o conteúdo do arquivo Excel em um DataFrame usando o pandas
        dados = pd.read_excel(caminho_arquivo)

        # Adiciona esse DataFrame à lista de dataframes
        dataframes.append(dados)

    # Combina todos os DataFrames da lista em um único DataFrame
    # 'ignore_index=True' serve para reorganizar os índices de forma contínua no novo DataFrame combinado
    dados_combinados = pd.concat(dataframes, ignore_index=True)

    # Define o nome (e caminho, se desejado) do arquivo de saída no formato CSV
    caminho_saida_csv = 'dados_2012_2022.csv'

    # Salva o DataFrame combinado como um arquivo CSV
    # 'index=False' evita que os índices sejam salvos como uma coluna extra no arquivo final
    dados_combinados.to_csv(caminho_saida_csv, index=False)

    # Mensagem de sucesso informando que o arquivo CSV foi criado com sucesso
    print(f'Arquivos combinados e salvos em {caminho_saida_csv}')

else:
    # Caso não haja pelo menos dois arquivos XLSX no diretório, exibe uma mensagem de aviso
    print('Pelo menos dois arquivos XLSX são necessários para combinar.')
```
